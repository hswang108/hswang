```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
 
infos_title_Builders = []
# range(0,2860,20):
for limitstart in range(0,100,20):
    url = 'https://www.myhousing.com.tw/index.php?option=com_kunena&view=category&catid=8&Itemid=1330&limitstart=%s' %limitstart
    print(url)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html5lib")

    for d in soup.find_all('tr', class_="category"):

        info_title_Builders = {}
        info_title_Builders['中文名稱'] = d.find('div', class_="krow").find('a',class_='hasTooltip topictitle').text.replace(' ', '').replace('\n', '') 
        info_title_Builders['作者'] = d.find('span', class_="kwho-user hasTooltip").text.replace(' ', '').replace('\n', '')                
        info_title_Builders['時間'] = d.find('span', class_="datepost").text.replace('主題時間 ： ', '') 
        info_title_Builders['回覆'] = d.find('div',class_="replies").find('span', class_="repliesnum").text.replace(' ', '').replace('\n', '') 
        info_title_Builders['檢視'] = d.find('div',class_="views").find('span', class_="viewsnum").text.replace(' ', '').replace('\n', '')
        
        
        
        
        infos_title_Builders.append(info_title_Builders)
print(infos_title_Builders)
```

```python
df_title_Builders = pd.DataFrame(infos_title_Builders)
df_title_Builders.to_csv('建商推案主旨.csv', encoding='utf_8_sig')

df_title_Builders
```
